```python
import RPi.GPIO as GPIO
from operator import mod
from time import sleep
import serial

# GPIO Setting, GPIO0Oput=1 when setting 16 Channels
#GPIO.setmode(GPIO.BCM)
#led=18
#GPIO.setup(led,GPIO.OUT)


Channel=["#A00C10D","#A00C11D","#A00C12D","#A00C13D","#A00C14D","#A00C15D","#A00C16D","#A00C17D","#A00C18D","#A00C19D","#A00C1AD","#A00C1BD","#A00C1CD","#A00C1DD","#A00C1ED","#A00C1FD"]
vset=[];NewChannel=[];V=[];DeltaV=[];VNew=[];TotalCommand=[];NewChannelc=[];Vc=[];DeltaVc=[];VNewc=[];TotalCommandc=[]

# Open file
fname = open('voltage.txt', 'r')      

for v in iter(fname):
    v=v.strip()                      # 去掉頭尾空白
    vset.append(v)

fname.close()


LoopN=int(vset[0])       # Set Loop Number
Sleep=float(vset[1])     # Set dt

#-------------------------------Ch1 to Ch16
for i in range(2,34):
    if i < 18:
        x=vset[i]
        V.append(float(x))
    else:
        y=vset[i]
        DeltaV.append(float(y))

for x in range(0,1):
    for i in range(0,16):
        NewChannel.append(Channel[i])
        
        if DeltaV[i]==0:
           V1="%04x" % round((V[i])*32.7675)                # 65535/2000=32.7675
           VNew.append(V1)
           TotalCommand.append(Channel[i]+V1+'\n')
        else:
           V1="%04x" % round((V[i]+x*DeltaV[i])*32.7675)           # 65535/2000=32.7675
           VNew.append(V1)   
           TotalCommand.append(Channel[i]+V1+'\n')      
        
print(V)
print(DeltaV)
#print(TotalCommand)
#--------------------------------------------------------------------------

#-------------------------------Ch17 to Ch32
for i in range(34,66):
    if i < 50:
        x1=vset[i]
        Vc.append(float(x1))
    else:
        y1=vset[i]
        DeltaVc.append(float(y1))

for x in range(0,LoopN):
    for i in range(0,16):
        NewChannelc.append(Channel[i])
        
        if DeltaVc[i]==0:
           V1="%04x" % round((Vc[i])*32.7675)                # 65535/2000=32.7675
           VNewc.append(V1)
           TotalCommandc.append(Channel[i]+V1+'\n')
        else:
           V1="%04x" % round((Vc[i]+x*DeltaVc[i])*32.7675)           # 65535/2000=32.7675
           VNewc.append(V1)   
           TotalCommandc.append(Channel[i]+V1+'\n')              
        
print(Vc)
print(DeltaVc)
#print(TotalCommandc)
print('Voltage Setting is OK')
#--------------------------------------------------------------------------

#Open rs232 and Send Command for Channel1 to Channel 16, fix Voltage
ser=serial.Serial('/dev/ttyUSB0', 38400, timeout=0.1)
for num in range(0,len(TotalCommand)):
    Output=TotalCommand[num]
    

    if mod(num+1,16)==0:   #complete one cycle, turn on led
        #GPIO.output(led, True)
        ser.write(Output.encode())
        #GPIO.output(led, False)
        #print(num+1)
        print(Output)
        #sleep(Sleep)
        #print('Finish One Loop')
    else:
        #GPIO.output(led, False)
        #print(num+1)
        ser.write(Output.encode())
        #print(Output)
ser.close()

#-----------------------------------------------------------------------------------

#Open rs232 and Send Command, Channel17 to Channel 32, Voltage Scan
ser=serial.Serial('/dev/ttyUSB1', 38400, timeout=0.1)
for num in range(0,len(TotalCommandc)):
    Output=TotalCommandc[num]
    

    if mod(num+1,16)==0:   #complete one cycle, turn on led
        #GPIO.output(led, True)
        ser.write(Output.encode())
        #GPIO.output(led, False)
        #print(num+1)
        print(Output)
        sleep(Sleep)
        #print('Finish One Loop')
    else:
        #GPIO.output(led, False)
        #print(num+1)
        ser.write(Output.encode())
        #print(Output)
ser.close()

#-----------------------------------------------------------------------------------
```
